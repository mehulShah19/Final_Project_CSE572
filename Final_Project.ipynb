```python
removedColumns = 0
for i in range(0,len(column_list)):
    columnLeft = column_list[i]
    for j in range(i+1,len(column_list)):
        columnRight = column_list[j]
        correlatedValue = X_train_df[columnLeft].corr(X_train_df[columnRight])
        if(correlatedValue >= 0.8 or correlatedValue <= -0.8):
            X_train_df = X_train_df.drop(columnLeft,1)
            removedColumns += 1
            print(columnLeft)
            break
print(removedColumns)
```
